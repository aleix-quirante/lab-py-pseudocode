### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# 1. Ask for the limit number (N)
# We use int() because you can't have "half" a loop iteration
n = int(input("Enter the number limit (N): "))

# 2. Initialize the accumulator (The Bus)
total_sum = 0

# 3. The Loop
# range(1, n + 1) generates numbers: 1, 2, 3 ... up to N
# (We put n + 1 because Python stops BEFORE the last number)
for i in range(1, n + 1):
    total_sum = total_sum + i

# 4. Print the final result
print(f"The sum of the first {n} numbers is: {total_sum}")
```

### Exercise 4: Find Maximum of Three Numbers

Write pseudocode that finds the maximum of three numbers entered by the user. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# 1. Capture the "input levels" (three numbers)
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

# 2. Comparison Logic
if num1 >= num2 and num1 >= num3:
    # If num1 is the "loudest"
    maximum = num1
elif num2 >= num1 and num2 >= num3:
    # If num2 is the "loudest"
    maximum = num2
else:
    # If neither 1 nor 2 are the highest, num3 wins by default
    maximum = num3

# 3. Final Output
print(f"The maximum number is: {maximum}")
```

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->

```python
# Data entry
text = input("Enter a sentence: ")

# 2. Filter about the serch
vowels = "aeiouAEIOU"

# 3. acomulator
vowel_count = 0

# 4. El bucle 
for char in text:
    if char in vowels:
        vowel_count += 1  # Si la letra es una vocal, aumentamos el contador

# 5. Resultado final
print(f"The number of vowels is: {vowel_count}")
```
